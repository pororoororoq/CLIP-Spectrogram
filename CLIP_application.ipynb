```python
import os
import shutil

# Define source directories (folders to extract from)
source_folders = ["for-norm", "for-rerec"]

# Define dataset paths (Change paths accordingly)
base_dataset_path = "/content/"
destination_root = "/content/dataset"

# Define dataset splits and categories
splits = ["training", "testing", "validation"]
categories = ["real", "fake"]

def restructure_dataset(base_path, output_root):
    """
    Restructure dataset by copying audio files from specific folders
    into a unified format for deepfake detection.
    """
    for source in source_folders:
        source_path = os.path.join(base_path, source, source)

        for split in splits:
            for category in categories:
                # Define source and destination paths
                src_dir = os.path.join(source_path, split, category)
                dst_dir = os.path.join(output_root, split, category)

                # Create target directory if it doesn't exist
                os.makedirs(dst_dir, exist_ok=True)

                # Copy files from source to target directory
                if os.path.exists(src_dir):  # Ensure source folder exists
                    for file in os.listdir(src_dir):
                        src_file = os.path.join(src_dir, file)
                        dst_file = os.path.join(dst_dir, file)

                        if os.path.isfile(src_file):  # Ensure it's a file
                            shutil.copy(src_file, dst_file)  # Copy without metadata

                    print(f"Copied {len(os.listdir(src_dir))} files from {src_dir} to {dst_dir}")

# Run the restructuring
restructure_dataset(base_dataset_path, destination_root)

print("Dataset restructuring complete!")
```

```python
# 1. Data Augmentation for Spectrograms
def augment_audio(waveform, sample_rate):
    # Ensure waveform is at least 2D (1, num_samples)
    if waveform.dim() == 1:
        waveform = waveform.unsqueeze(0)

    # Ensure waveform is consistent in shape
    if waveform.shape[0] > 1:  # If stereo, convert to mono
        waveform = torch.mean(waveform, dim=0, keepdim=True)

    # Add noise
    noise = torch.randn_like(waveform[:, :waveform.shape[1]]) * 0.01
    waveform += noise

    return waveform
```

```python
import torchaudio
import torch
import torch.nn.functional as F
import torchvision.transforms as T

def audio_to_mel_tensor(audio_path):
    try:
        waveform, sample_rate = torchaudio.load(audio_path)
        # Convert Stereo to Mono
        if waveform.shape[0] > 1:
            waveform = torch.mean(waveform, dim=0, keepdim=True)

        # Optional: Light data augmentation (noise)
        noise = torch.randn_like(waveform) * 0.01
        waveform += noise
        # (A) Possibly apply time masking or freq masking after mel:
        mel_spec = torchaudio.transforms.MelSpectrogram(...)(waveform)
        mel_db = torchaudio.transforms.AmplitudeToDB()(mel_spec)

        # Insert advanced augmentation here:
        time_mask = torchaudio.transforms.TimeMasking(time_mask_param=10)
        freq_mask = torchaudio.transforms.FrequencyMasking(freq_mask_param=5)
        mel_db = time_mask(mel_db)
        mel_db = freq_mask(mel_db)

        # MelSpectrogram
        mel_spec = torchaudio.transforms.MelSpectrogram(
        sample_rate=sample_rate,
        n_mels=64,   # Fewer bins
        n_fft=1024,  # Possibly smaller FFT size
        hop_length=256
    )(waveform)

        mel_db = torchaudio.transforms.AmplitudeToDB()(mel_spec)

        # Normalize the mel spectrogram to [0,1] or similar
        mel_db_norm = (mel_db - mel_db.min()) / (mel_db.max() - mel_db.min() + 1e-6)

        # mel_db_norm shape is [1, freq, time]. Make it 3 channels for CLIP (RGB).
        # Expand to shape [3, freq, time]
        mel_3c = mel_db_norm.repeat(3, 1, 1)
        mel_3c = F.interpolate(mel_3c.unsqueeze(0), size=(224, 224), mode='bilinear', align_corners=False).squeeze(0)
        return mel_3c  # Torch tensor, 3 x freq x time
    except RuntimeError as e:
        print(f"Error decoding audio file: {audio_path}. Error: {e}")
        # Return None or a placeholder tensor to skip this sample
        return None
```

```python
# 4. Multi-Level Feature Adaptation
class MultiLevelAdapter(nn.Module):
    def __init__(self, input_dim=512, output_dim=256):  # Input must match CLIP output
        super(MultiLevelAdapter, self).__init__()
        self.adapter1 = nn.Sequential(
            nn.Linear(input_dim, output_dim),  # 512 -> 256
            nn.ReLU(),
            nn.Linear(output_dim, output_dim)  # 256 -> 256
        )
        self.adapter2 = nn.Sequential(
            nn.Linear(output_dim, output_dim),
            nn.ReLU(),
            nn.Linear(output_dim, output_dim)
        )
        self.projection = nn.Linear(output_dim, input_dim)
        self.gamma = 0.1

    def forward(self, x):
        assert x.shape[1] == 512, f"Unexpected input shape: {x.shape}"
        x1 = self.adapter1(x)
        x2 = self.adapter2(x1)
        x2_projected = self.projection(x2)
        return self.gamma * x2_projected + (1 - self.gamma) * x

```

```python

# 5. CLIP Fine-Tuning with Multi-Level Feature Adaptation
def freeze_clip_weights(clip_model):
    for param in clip_model.parameters():
        param.requires_grad = False

# In your fine-tuner init:
class CLIPFineTuner(nn.Module):
    def __init__(self, clip_model):
        super(CLIPFineTuner, self).__init__()
        # Freeze original CLIP parameters to save memory
        freeze_clip_weights(clip_model)

        self.clip_model = clip_model.eval()  # also set to eval
        self.adapter = MultiLevelAdapter(input_dim=512, output_dim=256)
        self.fc = nn.Linear(512, 2)

    def forward(self, image):
        # With frozen CLIP, no gradient needed there
        with torch.no_grad():
            features = self.clip_model.encode_image(image)
        features = features.type(torch.float32)
        # The adapter + final FC are trainable
        adapted_features = self.adapter(features)
        output = self.fc(adapted_features)
        return output




fine_tuner = CLIPFineTuner(model).to(device)
optimizer = optim.Adam(fine_tuner.parameters(), lr=1e-4)
criterion = nn.CrossEntropyLoss()
```

```python
import gc
def train_model(model, train_loader, val_loader, epochs=10):
    best_val_loss = float("inf")
    for epoch in range(epochs):
        model.train()
        total_loss, total_correct = 0, 0
        for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            if -1 in labels:
                print("Skipping batch due to invalid label")
                continue
            valid_indices = labels != -1
            images = images[valid_indices].to(torch.float32).to(device)
            labels = labels[valid_indices].to(device)

            # Check if any valid samples remain after filtering
            if len(labels) == 0:
                print("Skipping batch as all samples have invalid labels")
                continue
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()

            for param in model.parameters():
                if param.grad is not None:
                    if torch.isnan(param.grad).any() or torch.isinf(param.grad).any():
                        print("NaN or inf gradients detected. Skipping gradient clipping.")
                        break  # Skip gradient clipping for this batch
            else:
                # Clip gradients only if they are valid
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()
            total_loss += loss.item()
            total_correct += (outputs.argmax(1) == labels).sum().item()
            gc.collect()
            torch.cuda.empty_cache()  # If using CUDA

        train_accuracy = total_correct / len(train_loader.dataset)
        avg_train_loss = total_loss / len(train_loader)
        print(f"Epoch {epoch+1}: Train Loss={avg_train_loss:.4f}, Train Acc={train_accuracy:.4f}")

train_model(fine_tuner, train_loader, val_loader, epochs=10)
```

```python
import torch
import torch.nn as nn
import torchaudio
import torch.nn.functional as F
import torchvision.transforms as T
import clip
from torch.utils.data import DataLoader
from tqdm import tqdm

device = "cuda" if torch.cuda.is_available() else "cpu"

###############################################################################
# LOAD CLIP (ViT-B/32) and optionally unfreeze some layers
###############################################################################
clip_model, clip_preprocess = clip.load("ViT-B/32", device=device)

def partial_freeze_clip_weights(clip_model, unfreeze_text=False, unfreeze_visual_layers=2):
    """
    Example function:
    - By default we freeze all text layers unless unfreeze_text=True
    - We unfreeze the last 'unfreeze_visual_layers' layers of the ViT visual transformer.
    """
    # 1) Freeze everything first
    for param in clip_model.parameters():
        param.requires_grad = False

    # 2) Optionally unfreeze text encoder
    #    (if you want to fine-tune text embeddings too — typically less common)
    if unfreeze_text:
        for name, param in clip_model.transformer.named_parameters():
            param.requires_grad = True
        for name, param in clip_model.token_embedding.named_parameters():
            param.requires_grad = True

    # 3) Unfreeze the last N layers of the ViT visual backbone
    #    Usually found in clip_model.visual.transformer.resblocks
    total_layers = len(clip_model.visual.transformer.resblocks)
    start_layer = max(0, total_layers - unfreeze_visual_layers)  # e.g., unfreeze 2 from the end
    for layer_idx in range(start_layer, total_layers):
        for name, param in clip_model.visual.transformer.resblocks[layer_idx].named_parameters():
            param.requires_grad = True

# Example: freeze text, partially unfreeze the last 2 layers of the visual trunk
partial_freeze_clip_weights(clip_model, unfreeze_text=False, unfreeze_visual_layers=2)

clip_model = clip_model.train()  # put in train mode (for dropout, LN stats, etc.)
clip_model.to(device)
```

```python
clip_transform = T.Compose([
    T.Resize((224, 224)),
    # T.CenterCrop(224), # optionally
    T.Normalize(mean=(0.48145466, 0.4578275, 0.40821073),
                std=(0.26862954, 0.26130258, 0.27577711)),
])
import torchvision.transforms as T
class AudioDataset(Dataset):
    def __init__(self, root_dir, subset, transform=None):
        self.root_dir = os.path.join(root_dir, subset)
        self.transform = transform
        self.data = []
        self.labels = []

        for label, category in enumerate(["real", "fake"]):
            category_dir = os.path.join(self.root_dir, category)
            for file in os.listdir(category_dir):
                if file.endswith(".wav"):
                    self.data.append(os.path.join(category_dir, file))
                    self.labels.append(label)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        audio_path = self.data[idx]
        label = self.labels[idx]
        waveform, sample_rate = torchaudio.load(audio_path)
        # Directly get a 3-channel mel-spectrogram as a tensor
        mel_tensor = waveform_to_clip_image_tensor(waveform)
        # Check if mel_tensor is None (due to decoding error)
        if mel_tensor is None:
            # Skip this sample and return a placeholder (or raise an error)
            print(f"Skipping sample: {audio_path}")
            return torch.zeros(3, 224, 224), -1  # Placeholder with a dummy label


        # Apply your CLIP transforms (resize, normalization)
        if self.transform is not None:
            mel_tensor = T.ToPILImage()(mel_tensor) # Convert the tensor to a PIL Image
            mel_tensor = self.transform(mel_tensor) # Apply CLIP's transforms



        return mel_tensor, label


transform = preprocess

subset_size = 500 #size will be subset_size * 2 because both fake and real exist
dataset_root = "dataset"
train_dataset = AudioDataset(root_dir=dataset_root, subset="training", transform=transform)
val_dataset = AudioDataset(root_dir=dataset_root, subset="validation", transform=transform)
#test_dataset = AudioDataset(root_dir=dataset_root, subset="testing", transform=transform)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, num_workers=0, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False, num_workers=0, pin_memory=True)
#test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False, num_workers=8)
```

```python
prompts = ["This is a fake audio clip", "This is a real audio clip"]
text_tokens = clip.tokenize(prompts).to(device)

# We'll keep these in a small nn.Parameter or buffer if you want them on GPU
with torch.no_grad():
    text_embeds = clip_model.encode_text(text_tokens)  # shape: [2, 512]
    text_embeds = text_embeds / text_embeds.norm(dim=-1, keepdim=True)
# Optionally requires_grad if we unfreeze the text side, else detach:
text_embeds = nn.Parameter(text_embeds, requires_grad=False)  # stored in param (frozen)
text_embeds = text_embeds.to(device)
```

```python
def waveform_to_clip_image_tensor(waveform, sr=16000):
    if waveform.dim() == 3 and waveform.shape[0] == 3:
    # It's already [3, H, W], so skip the mel transform.
      return waveform  # or handle differently
    # Ensure shape = [1, samples]
    if waveform.dim() == 1:
        waveform = waveform.unsqueeze(0)  # [1, samples]
    elif waveform.dim() == 3 and waveform.shape[0] == 1:
        waveform = waveform.squeeze(0)    # [1, samples]
    # If stereo: average across channels
    if waveform.shape[0] > 1:
        waveform = waveform.mean(dim=0, keepdim=True)

    mel_spec = torchaudio.transforms.MelSpectrogram(
        sample_rate=sr,
        n_mels=64,
        n_fft=1024,
        hop_length=256,
        pad=(1024 - 256)//2
    ).to(device)(waveform.to(device))

    # shape => [1, 64, time]
    mel_db = torchaudio.transforms.AmplitudeToDB()(mel_spec)

    mel_db_norm = (mel_db - mel_db.min()) / (mel_db.max() - mel_db.min() + 1e-6)
    # shape => [1, 64, time]

    mel_3c = mel_db_norm.repeat(3, 1, 1)  # shape => [3, 64, time]

    # Resize to [3, 224, 224]
    mel_3c = F.interpolate(
        mel_3c.unsqueeze(0), size=(224,224),
        mode="bilinear", align_corners=False
    ).squeeze(0)

    return mel_3c

clip_normalize = T.Normalize(
    mean=(0.48145466, 0.4578275, 0.40821073),
    std=(0.26862954, 0.26130258, 0.27577711)
)
```

```python
def train_model_with_text_embeds(clip_model, text_embeds, train_loader,
                                 epochs=5, lr=1e-5, device="cuda"):
    """
    - clip_model: your loaded CLIP model (with partial freeze or not)
    - text_embeds: shape [2, 512], index=0 => fake, index=1 => real
    - train_loader: yields (waveform, label) with label in {0,1}
    - We do a cross-entropy with dot products [batch_size, 2].
    """
    # We'll use a standard Adam optimizer on *whatever* is unfrozen in clip_model
    optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, clip_model.parameters()),
                                 lr=lr)
    criterion = nn.CrossEntropyLoss()

    # If text_embeds.requires_grad == True, also pass them to the optimizer.
    # But here we assume we froze them:
    text_embeds.requires_grad_(False)

    for epoch in range(epochs):
        clip_model.train()
        total_loss = 0.0
        total_samples = 0

        for waveforms, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            waveforms = waveforms.to(device)
            labels = labels.to(device)

            # Skip invalid label
            if (labels == -1).any():
                continue

            batch_size = waveforms.shape[0]
            # Convert each waveform -> mel [3,224,224]
            # We'll do it in a loop for clarity (you can vectorize if needed).
            image_tensors = []
            for i in range(batch_size):
                mel_3c = waveform_to_clip_image_tensor(waveforms[i])
                mel_3c = clip_normalize(mel_3c)
                image_tensors.append(mel_3c)
            # Stack => [batch_size, 3, 224, 224]
            image_batch = torch.stack(image_tensors, dim=0).to(device)

            # Forward pass: encode image
            image_emb = clip_model.encode_image(image_batch)  # shape [batch_size, 512]
            # L2-normalize
            image_emb = image_emb / image_emb.norm(dim=-1, keepdim=True)

            # text_embeds: [2, 512], so let's also ensure they're normalized
            # (Should already be normalized, but let's re-check):
            # text_embeds = text_embeds / text_embeds.norm(dim=-1, keepdim=True)
            # We'll skip re-normalizing if we did it once up front

            # Dot product => logits shape [batch_size, 2]
            # [batch_size, 512] x [512, 2] => [batch_size, 2]
            # We'll multiply by e.g. 100.0 to match CLIP's scale if we want.
            logits = 100.0 * image_emb @ text_embeds.T

            # Compute cross-entropy
            # labels in {0,1} => 0 => "fake", 1 => "real"
            loss = criterion(logits, labels)

            # Backprop
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            total_loss += loss.item() * batch_size
            total_samples += batch_size

        avg_loss = total_loss / total_samples
        print(f"Epoch {epoch+1}/{epochs} - Loss: {avg_loss:.4f}")

    print("Finished training with text embeddings!")
    return clip_model
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
import torchaudio
import numpy as np
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from tqdm import tqdm
from sklearn.metrics import roc_auc_score
import clip
import gc

# Define file structure
DATASET_ROOT = "/content/dataset"

# Binary class labels
CLASS_INDEX = {'Real': 1, 'Fake': 0}

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def setup_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

setup_seed(42)

def collate_fn(batch):
    """Custom collate function to handle variable-length spectrograms properly."""
    spectrograms, labels = zip(*batch)  # Unpack batch

    # Ensure all spectrograms have shape [3, H, W]
    max_H = max(s.shape[1] for s in spectrograms)
    max_W = max(s.shape[2] for s in spectrograms)

    # Initialize padded batch
    padded_spectrograms = torch.zeros(len(spectrograms), 3, max_H, max_W)

    for i, s in enumerate(spectrograms):
        _, h, w = s.shape
        padded_spectrograms[i, :, :h, :w] = s  # Pad each spectrogram correctly

    return padded_spectrograms, torch.tensor(labels, dtype=torch.float32)

# Define the dataset class
class AudioDeepfakeDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.data = []
        for label in ["real", "fake"]:
            label_dir = os.path.join(root_dir, label)
            for file in os.listdir(label_dir):
                self.data.append((os.path.join(label_dir, file), CLASS_INDEX[label.capitalize()]))

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        file_path, label = self.data[idx]
        try:
            waveform, sr = torchaudio.load(file_path)
        except RuntimeError as e:
            print(f"Error loading {file_path}: {e}")
            # Skip this sample:
            return torch.zeros(3, 224, 224), -1   # Return dummy data and label of the correct shape for your model


        # Convert to Mel Spectrogram
        mel_spectrogram = torchaudio.transforms.MelSpectrogram(sample_rate=sr,
          n_fft=512,  # Increase n_fft to capture more frequency info
          hop_length=256,
          n_mels=64  # Reduce n_mels to avoid empty filterbanks)
          )(waveform)
        mel_spectrogram = torch.log1p(mel_spectrogram)
        mel_spectrogram = (mel_spectrogram - mel_spectrogram.mean()) / mel_spectrogram.std()
        # Ensure spectrogram is [1, H, W]
        mel_spectrogram = mel_spectrogram.unsqueeze(0) if mel_spectrogram.ndim == 2 else mel_spectrogram

        # Normalize to 0-1 range
        mel_spectrogram = (mel_spectrogram - mel_spectrogram.min()) / (mel_spectrogram.max() - mel_spectrogram.min())

        # Convert single-channel to 3-channel (stacking)
        mel_spectrogram = mel_spectrogram.repeat(3, 1, 1)  # Convert [1, H, W] → [3, H, W]

        # Resize to CLIP's input size
        resize = transforms.Resize((224, 224))  # Resize to 224x224 for CLIP
        mel_spectrogram = resize(mel_spectrogram)

        return mel_spectrogram, label


# Load datasets
train_dataset = AudioDeepfakeDataset(os.path.join(DATASET_ROOT, "training"))
test_dataset = AudioDeepfakeDataset(os.path.join(DATASET_ROOT, "testing"))

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, num_workers=4, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False, num_workers=4, collate_fn=collate_fn)

# Load CLIP model with Adapters
class ClipAdapter(nn.Module):
    def __init__(self, c_in=1024, bottleneck=1024):
        super(ClipAdapter, self).__init__()
        self.fc1 = nn.Sequential(
            nn.Linear(c_in, bottleneck, bias=False),
            nn.LeakyReLU(inplace=False)
        )
        self.fc2 = nn.Sequential(
            nn.Linear(bottleneck, c_in, bias=False),
            nn.LeakyReLU(inplace=False)
        )

        self.fc1.to(device)
        self.fc2.to(device)


    def forward(self, x):
        x = self.fc1(x)
        y = self.fc2(x)
        return x, y

class CLIPWithAdapters(nn.Module):
    def __init__(self, clip_model, features):
        super().__init__()
        self.clip_model = clip_model
        self.visual_encoder = clip_model.visual
        self.transformer = clip_model.visual.transformer  # Access CLIP's transformer
        self.features = features
        # Add adapters at selected transformer layers:
        self.adapters = nn.ModuleList([ClipAdapter(1024, bottleneck=512) for _ in range(len(features))])

        # **Add a text projection layer from 768 -> 1024.**
        # (For ViT-L/14, the raw text features are 768-d.)
        self.text_projection = nn.Linear(768, 1024).to(device)
        self.dropout = nn.Dropout(0.2)
    def forward(self, x):
        # (Keep your image forward pass as before.)
        x = self.visual_encoder.conv1(x)  # Initial convolution
        x = x.reshape(x.shape[0], x.shape[1], -1).permute(0, 2, 1)
        x = torch.cat([
            self.visual_encoder.class_embedding.to(x.dtype) +
            torch.zeros(x.shape[0], 1, x.shape[-1], dtype=x.dtype, device=x.device),
            x
        ], dim=1)
        x = x + self.visual_encoder.positional_embedding.to(x.dtype)
        x = self.visual_encoder.ln_pre(x)

        for i, block in enumerate(self.transformer.resblocks):
            x = block(x)  # Pass through CLIP’s transformer block
            if i in self.features:
                adapted, residual = self.adapters[self.features.index(i)](x)
                if adapted.shape[-1] != x.shape[-1]:
                    adapted = torch.nn.functional.pad(adapted, (0, x.shape[-1] - adapted.shape[-1]))
                if residual.shape[-1] != x.shape[-1]:
                    residual = torch.nn.functional.pad(residual, (0, x.shape[-1] - residual.shape[-1]))
                x = 0.8 * x + 0.1 * adapted + 0.1 * residual
        x = self.dropout(x)
        x = self.visual_encoder.ln_post(x[:, 0, :])
        return x

    def encode_text(self, tokens):
        # Get raw text features (likely 768-dim) from the underlying model.
        raw_text_features = self.clip_model.encode_text(tokens)
        # Now project them to 1024 dimensions.
        text_features = self.text_projection(raw_text_features)
        return text_features

    def encode_image(self, image):
        return self.forward(image)



import open_clip  # Use OpenCLIP instead of OpenAI CLIP

clip_model, _, preprocess = open_clip.create_model_and_transforms(
    "ViT-L-14",
    pretrained="openai",
    device=device
)

clip_model = CLIPWithAdapters(clip_model, features=[4, 8, 12, 16, 20])
for i, block in enumerate(clip_model.clip_model.visual.transformer.resblocks[-4:]):
    for param in block.parameters():
        param.requires_grad = True
clip_model.train()


def encode_text_with_prompt_ensemble(model, label, device):
    prompts = [
        "A real audio sample" if label == "Real" else "A fake audio sample",
        "A normal sound clip" if label == "Real" else "Anomalous sound detected",
        "Authentic recording" if label == "Real" else "Digitally manipulated speech"
    ]
    tokens = clip.tokenize(prompts).to(device)
    with torch.no_grad():
        # This call now uses your updated encode_text, so the output will be [N, 1024]
        text_features = model.encode_text(tokens)
        print("Projected text features shape:", text_features.shape)  # Should print (3, 1024)
    return text_features.mean(dim=0).unsqueeze(0)  # Returns shape [1, 1024]




text_feature_list = []
with torch.no_grad():
    for label in CLASS_INDEX.keys():
        text_feature = encode_text_with_prompt_ensemble(clip_model, label, device)
        text_feature_list.append(text_feature)


class WeightedBCELoss(nn.Module):
    def __init__(self, weight=0.75):  # Adjust weight if needed
        super().__init__()
        self.bce = nn.BCEWithLogitsLoss(pos_weight=torch.tensor([weight], device=device))

    def forward(self, predictions, targets):
                return self.bce(predictions, targets)

# Define Training Loop
def train_model(model, train_loader, criterion, optimizer, epochs=50):
    model.train()
    for epoch in range(epochs):
        loss_list = []
        correct_predictions = 0
        total_samples = 0

        for spectrogram, label in tqdm(train_loader):
            # Skip batches with invalid samples
            if torch.isnan(label).any() or (label == -1).any():
                print("Skipping batch containing invalid sample(s)")
                continue

            spectrogram = spectrogram.to(device)
            label = label.to(device)  # labels are 0/1 (float)

            optimizer.zero_grad()
            with torch.cuda.amp.autocast():
                # Get image (audio) features; expect shape [B, 1024]
                spectrogram_features = model.encode_image(spectrogram)

                # Expand the stored text features to the batch size.
                # text_feature_list[CLASS_INDEX['Real']] and text_feature_list[CLASS_INDEX['Fake']]
                # should now both be of shape [1, 1024]
                real_text = text_feature_list[CLASS_INDEX['Real']].expand(spectrogram_features.size(0), -1)
                fake_text = text_feature_list[CLASS_INDEX['Fake']].expand(spectrogram_features.size(0), -1)
                # Choose the text feature per sample based on the label.
                batch_text_features = torch.where(label.unsqueeze(1) == 1, real_text, fake_text)
                # Compute the dot product (raw logits) per sample.
                anomaly_score = 100.0 * (spectrogram_features * batch_text_features).sum(dim=1)
                # Do NOT apply sigmoid here since BCEWithLogitsLoss expects logits.

                # For prediction, apply sigmoid
                predicted_probs = torch.sigmoid(anomaly_score)
                predicted_labels = (predicted_probs > 0.5).int()

                correct_predictions += (predicted_labels == label.int()).sum().item()
                total_samples += label.size(0)

                # Compute loss with raw logits
                loss = criterion(anomaly_score, label.float())

            if torch.isnan(loss):
                print("Loss is NaN. Skipping this batch.")
                continue

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.5)
            optimizer.step()

            loss_list.append(loss.item())

        gc.collect()
        torch.cuda.empty_cache()
        accuracy = correct_predictions / total_samples if total_samples else 0
        print(f"Epoch {epoch+1}: Loss = {np.mean(loss_list):.4f}, Accuracy = {accuracy:.4f}")
    torch.save(model, 'CLIP0131_0510.pth')


# Define Zero-Shot Evaluation
def evaluate_zero_shot(model, test_loader, text_features):
    model.eval()
    scores = []
    labels = []
    with torch.no_grad():
        for spectrogram, label in tqdm(test_loader):
            spectrogram = spectrogram.to(device)
            label = label.to(device)
            spectrogram_features = model.encode_image(spectrogram)
            anomaly_score = (100.0 * spectrogram_features @ text_features.T).squeeze()
            scores.append(anomaly_score.cpu().numpy())
            labels.append(label.cpu().numpy())
    auc = roc_auc_score(np.concatenate(labels), np.concatenate(scores))
    print(f"Zero-Shot Test AUC: {auc:.4f}")
    return auc

# Train the model
train_model(clip_model, train_loader, WeightedBCELoss(), optim.Adam(clip_model.parameters(), lr=0.0001), epochs=5)

# Evaluate Zero-Shot Performance
evaluate_zero_shot(clip_model, test_loader, text_feature_list[CLASS_INDEX['Fake']])

```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
import torchaudio
import numpy as np
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from tqdm import tqdm
from sklearn.metrics import roc_auc_score
import torch.nn.functional as F
import gc
# Define file structure
DATASET_ROOT = "/content/dataset"

# Binary class labels
CLASS_INDEX = {'Real': 1, 'Fake': 0}

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define the dataset class
class AudioDeepfakeDataset(Dataset):
    def __init__(self, root_dir, transform=None, desired_length = 25000):
        self.root_dir = root_dir
        self.transform = transform
        self.data = []
        self.desired_length = desired_length
        for label in ["real", "fake"]:
            label_dir = os.path.join(root_dir, label)
            for file in os.listdir(label_dir):
                self.data.append((os.path.join(label_dir, file), CLASS_INDEX[label.capitalize()]))

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        file_path, label = self.data[idx]
        try:
            waveform, sr = torchaudio.load(file_path)
        except RuntimeError as e:
            print(f"Error loading {file_path}: {e}")
            # Skip this sample:
            return torch.zeros(1, self.desired_length), -1   # Return dummy data and label

        # Pad or trim waveforms to a fixed length

        # Pad if shorter
        if waveform.shape[1] < self.desired_length:
            waveform = F.pad(waveform, (0, self.desired_length - waveform.shape[1]))
        # Trim if longer
        else:
            waveform = waveform[:, :self.desired_length]

        if self.transform:
            waveform = self.transform(waveform)
        return waveform, label

# Define the CNN-LSTM model
class CNN_LSTM(nn.Module):
    def __init__(self, num_classes=2, desired_length=25000):
        super(CNN_LSTM, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=16, kernel_size=5, stride=1, padding=2)
        self.conv2 = nn.Conv1d(in_channels=16, out_channels=32, kernel_size=5, stride=1, padding=2)
        self.pool = nn.MaxPool1d(kernel_size=2, stride=2)

        # The input size for LSTM should be the number of features (channels)
        # after the CNN layers, which is 32 in this case.
        self.lstm_input_size = 32

        # Define the LSTM layer with the correct input size
        self.lstm = nn.LSTM(input_size=self.lstm_input_size, hidden_size=64, num_layers=2, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(64 * 2, num_classes)  # *2 for bidirectional

    def forward(self, x):
        x = x.type(torch.float32)
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))

        # Reshape for LSTM (batch, sequence, features)
        x = x.permute(0, 2, 1)  # Transpose to [batch_size, sequence_length, num_features]

        # Pass through LSTM and get output
        # self.lstm.flatten_parameters() # Optional: to improve memory efficiency
        x, _ = self.lstm(x)

        # Take the output from the last time step
        x = self.fc(x[:, -1, :])
        return x


# Load datasets
train_dataset = AudioDeepfakeDataset(os.path.join(DATASET_ROOT, "training"))
test_dataset = AudioDeepfakeDataset(os.path.join(DATASET_ROOT, "validation"))

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, num_workers=4, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False, num_workers=4, pin_memory=True)

# Initialize model, loss, and optimizer
model = CNN_LSTM(num_classes=2).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)

# Train the model
def train_model(model, train_loader, criterion, optimizer, epochs=10):
    model.train()
    for epoch in range(epochs):
        loss_list = []
        for waveform, label in tqdm(train_loader):
            if -1 in label:
                print("Skipping batch containing invalid sample(s)")
                continue  # Skip this batch
            waveform = waveform.to(device)
            label = label.long().to(device)

            if len(label) == 0:
                print("Skipping batch as all samples have invalid labels")
                continue
            optimizer.zero_grad()
            output = model(waveform)
            loss = criterion(output, label)
            loss.backward()

            # Gradient clipping (if no NaN/inf gradients)
            if not any(torch.isnan(p.grad).any() or torch.isinf(p.grad).any()
                       for p in model.parameters() if p.grad is not None):
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            else:
                print("NaN or inf gradients detected. Skipping gradient clipping.")

            optimizer.step()
            loss_list.append(loss.item())

            # Clear CUDA cache more selectively (e.g., every few batches)
            if len(loss_list) % 10 == 0:  # Adjust frequency as needed
              gc.collect()
              torch.cuda.empty_cache()

        print(f"Epoch {epoch+1}: Loss = {np.mean(loss_list)}")


# Evaluate the model
def evaluate_model(model, test_loader):
    model.eval()
    scores = []
    labels = []
    with torch.no_grad():
        for waveform, label in tqdm(test_loader):
            waveform = waveform.to(device)
            label = label.to(device)
            output = model(waveform)
            scores.append(torch.softmax(output, dim=1)[:, 1].cpu().numpy())
            labels.append(label.cpu().numpy())
    auc = roc_auc_score(np.concatenate(labels), np.concatenate(scores))
    print(f"Test AUC: {auc:.4f}")
    return auc

# Train and evaluate
#train_model(model, train_loader, criterion, optimizer, epochs=10)
#valuate_model(model, test_loader)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
model = torch.load("cnn_deepfake_audio.pth")
model.eval()
def evaluate_zero_shot(model, zero_shot_loader):
    model.eval()
    scores = []
    labels = []

    with torch.no_grad():
        for waveform, label in tqdm(zero_shot_loader):
            waveform = waveform.to(device)
            label = label.to(device)

            # Pass waveform through CNN model to get predictions
            outputs = model(waveform)

            # Get the probability of the 'Fake' class
            score = outputs[:, 1].cpu().numpy()  # Assuming 'Fake' is class 0

            scores.extend(score)  # Use extend to add elements to the list
            labels.extend(label.cpu().numpy())

    # Calculate ROC curve and AUC
    fpr, tpr, thresholds = roc_curve(labels, scores)
    roc_auc = auc(fpr, tpr)

    # Plot ROC curve
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Zero-Shot ROC (CNN-LSTM)')
    plt.legend(loc="lower right")
    plt.show()

    print(f'Zero-Shot Test AUC: {roc_auc:.4f}')
    return roc_auc

# Call evaluate_zero_shot with the loaded CNN model
zero_shot_auc = evaluate_zero_shot(model, zero_shot_loader)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
import torchvision.transforms as T
from tqdm import tqdm
import torchaudio
import torch
import torch.nn.functional as F
import tempfile
import numpy as np
import os
import io

# (Reuse your audio_to_mel_tensor_for_test function)
def audio_to_mel_tensor_for_test(audio_bytes_io):
    try:
        # Create a temporary WAV file
        with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as temp_file:
            temp_file_path = temp_file.name
            waveform_tensor = torch.tensor(np.frombuffer(audio_bytes_io.read(), dtype=np.float32))
            torchaudio.save(temp_file_path, waveform_tensor.unsqueeze(0), sample_rate=16000)

        waveform, sample_rate = torchaudio.load(temp_file_path)
        os.remove(temp_file_path)

        # Convert stereo to mono if necessary
        if waveform.shape[0] > 1:
            waveform = torch.mean(waveform, dim=0, keepdim=True)

        # (Optional) No noise in few-shot evaluation—if training didn’t include it
        # noise = torch.randn_like(waveform) * 0.01
        # waveform += noise

        # Compute mel spectrogram and convert to dB scale
        mel_spec = torchaudio.transforms.MelSpectrogram(
            sample_rate=sample_rate,
            n_mels=64,
            n_fft=1024,
            hop_length=256
        )(waveform)
        mel_db = torchaudio.transforms.AmplitudeToDB()(mel_spec)
        mel_db_norm = (mel_db - mel_db.min()) / (mel_db.max() - mel_db.min() + 1e-6)
        # Repeat channels to create a 3‑channel “image”
        mel_3c = mel_db_norm.repeat(3, 1, 1)
        # Resize to 224x224 (assuming CLIP input size)
        mel_3c = F.interpolate(mel_3c.unsqueeze(0), size=(224, 224), mode='bilinear', align_corners=False).squeeze(0)
        return mel_3c
    except RuntimeError as e:
        print(f"Error decoding audio file: {audio_bytes_io}. Error: {e}")
        return None

def evaluate_few_shot(model, support_loader, query_loader, device='cuda'):
    """
    Few-shot evaluation: compute prototypes from support set and evaluate on query set.

    Assumptions:
      - support_loader and query_loader yield (waveform_batch, label_batch).
      - Labels are 0 (fake) or 1 (real).
      - The audio preprocessing in the support and query loaders is not applied;
        we compute the mel spectrogram using audio_to_mel_tensor_for_test.
    """
    model.eval()

    # Define the same transformation as used during zero-shot evaluation.
    transform = T.Compose([
        T.Resize((224, 224)),
        T.Normalize(mean=(0.48145466, 0.4578275, 0.40821073),
                    std=(0.26862954, 0.26130258, 0.27577711)),
    ])

    # --- Step 1. Compute Prototypes from the Support Set ---
    # We'll accumulate embeddings for each class.
    prototypes = {0: [], 1: []}

    print("Computing prototypes from support set...")
    with torch.no_grad():
        for waveform_batch, label_batch in tqdm(support_loader, desc="Support Set"):
            for waveform, label in zip(waveform_batch, label_batch):
                # Convert the waveform to a bytes stream and then to a mel tensor
                waveform_np = waveform.cpu().numpy().flatten().astype(np.float32)
                waveform_bytes = waveform_np.tobytes()
                audio_io = io.BytesIO(waveform_bytes)
                mel_tensor = audio_to_mel_tensor_for_test(audio_io)
                if mel_tensor is None:
                    continue
                mel_tensor = transform(mel_tensor).unsqueeze(0).to(device)  # Shape: [1, 3, 224, 224]

                # Obtain the image (audio) embedding.
                # Use your model’s image encoder consistently.
                image_emb = model.clip_model.encode_image(mel_tensor)
                # Normalize the embedding
                image_emb = image_emb / image_emb.norm(dim=-1, keepdim=True)
                # Save the embedding (squeeze the batch dimension)
                prototypes[int(label.item())].append(image_emb.squeeze(0))

    # Check that we have support examples for each class.
    if not prototypes[0] or not prototypes[1]:
        print("Not enough support examples for one or both classes.")
        return None

    # Average the embeddings for each class and normalize.
    proto_fake = torch.stack(prototypes[0]).mean(dim=0, keepdim=True)
    proto_real = torch.stack(prototypes[1]).mean(dim=0, keepdim=True)
    proto_fake = proto_fake / proto_fake.norm(dim=-1, keepdim=True)
    proto_real = proto_real / proto_real.norm(dim=-1, keepdim=True)

    print("Prototypes computed.")

    # --- Step 2. Evaluate on the Query Set ---
    scores = []  # Decision scores (e.g., sim(real) - sim(fake))
    labels  = []  # Ground truth labels

    print("Evaluating query set...")
    with torch.no_grad():
        for waveform_batch, label_batch in tqdm(query_loader, desc="Query Set"):
            for waveform, label in zip(waveform_batch, label_batch):
                waveform_np = waveform.cpu().numpy().flatten().astype(np.float32)
                waveform_bytes = waveform_np.tobytes()
                audio_io = io.BytesIO(waveform_bytes)
                mel_tensor = audio_to_mel_tensor_for_test(audio_io)
                if mel_tensor is None:
                    continue
                mel_tensor = transform(mel_tensor).unsqueeze(0).to(device)
                image_emb = model.clip_model.encode_image(mel_tensor)
                image_emb = image_emb / image_emb.norm(dim=-1, keepdim=True)

                # Compute cosine similarities with both prototypes.
                sim_fake = F.cosine_similarity(image_emb, proto_fake)
                sim_real = F.cosine_similarity(image_emb, proto_real)

                # Option: use the difference between the similarities as a decision score.
                score = (sim_real - sim_fake).item()
                scores.append(score)
                labels.append(label.cpu().item())

    # --- Step 3. Compute ROC Curve and AUC ---
    fpr, tpr, _ = roc_curve(labels, scores, pos_label=1)
    roc_auc = auc(fpr, tpr)

    # Plot ROC curve.
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate (K=20)')
    plt.ylabel('True Positive Rate')
    plt.title('Few-Shot ROC')
    plt.legend(loc="lower right")
    plt.show()

    print(f"Few-Shot Test AUC: {roc_auc:.4f}")
    return roc_auc

few_shot_auc = evaluate_few_shot(model_clip, support_loader, query_loader, device)
```

```python
import torch.nn as nn
import torch.optim as optim
import numpy as np
from tqdm import tqdm
import gc

def few_shot_train(model, support_loader, num_epochs=10, lr=0.001):
    # Freeze all layers except the final classifier (fc)
    for name, param in model.named_parameters():
        if "fc" not in name:
            param.requires_grad = False
        else:
            param.requires_grad = True

    optimizer = optim.Adam(model.fc.parameters(), lr=lr)
    criterion = nn.CrossEntropyLoss()

    model.train()
    for epoch in range(num_epochs):
        loss_list = []
        for waveform, label in tqdm(support_loader, desc=f"Support Set Epoch {epoch+1}"):
            # Skip any invalid samples (if label == -1)
            if -1 in label:
                continue

            waveform = waveform.to(device)
            label = label.long().to(device)

            optimizer.zero_grad()
            output = model(waveform)  # forward pass
            loss = criterion(output, label)
            loss.backward()
            optimizer.step()
            loss_list.append(loss.item())

        avg_loss = np.mean(loss_list) if loss_list else float('nan')
        print(f"Few-Shot Epoch {epoch+1}: Loss = {avg_loss:.4f}")
        gc.collect()
        torch.cuda.empty_cache()

    return model

# Fine-tune your model on the support set
few_shot_model = few_shot_train(model, support_loader, num_epochs=10, lr=0.001)
```
